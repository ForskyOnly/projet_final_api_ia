```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("/home/utilisateur/Documents/dev/devia/projet_final_api_ia/database/data/cars.csv")
df
```

### Suppression de la colonne Car_ID 

```python
df = df.drop("car_ID", axis=1)
```

### Creation de colonne "model" pour les marques de voitures

```python
df["model"] = df["CarName"].apply(lambda x: x.split(" ", 1)[1] if len(x.split(" ", 1)) > 1 else x.split(" ", 1)[0])
```

### Nettoyage de la colonne "CarName"

```python
df["CarName"] = df["CarName"].apply(lambda x: x.split(" ")[0])
```

```python
df
```

## Nettoyage  des colonnes 

```python
# Nettoyage colonne CarName, on remplace les fautes et les valeurs incorrect par des valeurs correct
df["CarName"] = df["CarName"].replace("porcshce", "porsche")
df["CarName"] = df["CarName"].replace("alfa-romero", "alfa-romeo")
df["CarName"] = df["CarName"].replace("vw", "volkswagen")
df["CarName"] = df["CarName"].replace("vokswagen", "volkswagen")
df["CarName"] = df["CarName"].replace("toyouta", "toyota")
df["CarName"] = df["CarName"].replace("Nissan", "nissan")
df["CarName"] = df["CarName"].replace("maxda", "mazda")
df["enginetype"] = df["enginetype"].replace("ohcf", "ohc") # moteur inexistant 
df["enginetype"] = df["enginetype"].replace("dohcv", "dohc") # moteur inexistant 

df["doornumber"] = df["doornumber"].replace("two", 2)
df["doornumber"] = df["doornumber"].replace("four", 4)

df["cylindernumber"] = df["cylindernumber"].replace("two", 2)
df["cylindernumber"] = df["cylindernumber"].replace("three", 3)
df["cylindernumber"] = df["cylindernumber"].replace("four", 4)   
df["cylindernumber"] = df["cylindernumber"].replace("five", 5)
df["cylindernumber"] = df["cylindernumber"].replace("six", 6)
df["cylindernumber"] = df["cylindernumber"].replace("eight", 8)
df["cylindernumber"] = df["cylindernumber"].replace("twelve", 12)
```

### Convertion des valeurs en métrique

```python
df['wheelbase'] = df['wheelbase'].apply(lambda x: x * 2.54).round(2) # 1 pouce = 2,54 centimètres.
df['carlength'] = df['carlength'].apply(lambda x: x * 2.54).round(2)
df['carwidth'] = df['carwidth'].apply(lambda x: x * 2.54).round(2)
df['carheight'] = df['carheight'].apply(lambda x: x * 2.54).round(2)
df['curbweight'] = df['curbweight'].apply(lambda x: x * 0.45359237).round(2) # 1 livre = 0,45359237 kilogramme

```

```python
df.to_csv("/home/utilisateur/Documents/dev/devia/projet_final_api_ia/database/data/cleaned_data.csv", index=False)
```
