```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/utilisateur/Documents/dev/devia/projet_final_api_ia/database/data/cleaned_data.csv")
```

```python
df
```

```python
cor = df.corr()
print(cor['price'].sort_values(ascending=False))
```

```python
plt.figure(figsize=(20, 8))
sns.heatmap(cor, cmap='YlGnBu', annot=True)
plt.show()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
sns.regplot(y='curbweight', x='price', data=df, ax=axes[0])
sns.regplot(y='horsepower', x='price', data=df, ax=axes[1])
sns.regplot(y='enginesize', x='price', data=df, ax=axes[2])
sns.regplot(y='boreratio', x='price', data=df, ax=axes[3])
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

sns.regplot(y='wheelbase', x='price', data=df, ax=axes[0])
sns.regplot(y='carlength', x='price', data=df, ax=axes[1])
sns.regplot(y='carwidth', x='price', data=df, ax=axes[2])
sns.scatterplot(y='fuelsystem', x='price', data=df, ax=axes[3])
```
